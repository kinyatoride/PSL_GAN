```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ser = pd.Series([314.812, 150.19996643066406],
                index=['Original', 'New'])
```

```python
ser_batch = pd.Series([150.19996643066406,
                       78.5765151977539,
                       38.994014739990234,
                       21.851295471191406], 
                      index=[32, 64, 128, 256])
```

```python
fig, ax = plt.subplots(figsize=(2,3))
ser.plot(kind='bar', zorder=2)
ax.grid(zorder=1)
ax.set(ylabel='1 epoch time (ms)')
plt.xticks(rotation=0)
```

```python
fig, ax = plt.subplots(figsize=(4,4))
ser_batch.plot(kind='bar', zorder=2)
ax.grid(zorder=1)

ax.set(ylabel='1 epoch time (ms)',
       xlabel='batch size')
plt.xticks(rotation=0)
```
